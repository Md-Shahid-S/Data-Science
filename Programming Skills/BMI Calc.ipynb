BMI Calculator

```python
import matplotlib.pyplot as plt


def plot_bmi(bmi):
    """Plot BMI results."""
    categories = ["Underweight", "Normal weight", "Overweight", "Obesity"]
    thresholds = [18.5, 24.9, 29.9, 40] 
    values = [bmi if bmi < t else t for t in thresholds]  

    plt.bar(categories, thresholds, color='lightgrey', label='Category Thresholds')
    plt.bar(categories, values, color='blue', label='Your BMI')

    plt.axhline(bmi, color='red', linestyle='--', label=f'Your BMI: {bmi:.2f}')

    plt.xlabel("BMI Categories")
    plt.ylabel("BMI Values")
    plt.title("BMI Categories and Your BMI")
    plt.legend()
    plt.show()
```

```python
def save_results_to_file(Name, bmi, category, weight, height_cm):
    """Save BMI results to a file."""
    with open("bmi_results.txt", "a") as file:
        file.write("-" * 30 + "\n")
        file.write(f"Name: {Name} \n")
        file.write(f"Weight: {weight} kg\n")
        file.write(f"Height: {height_cm} cm\n")
        file.write(f"BMI: {bmi:.2f}\n")
        file.write(f"Category: {category}\n")
        file.write("-" * 30 + "\n")
    print("Results saved to 'bmi_results.txt'")
    print("-" * 30)
```

```python
# Input data
name = input("Enter your name: ")
height = float(input("Enter your height in cm: "))
weight = float(input("Enter your weight in kg: "))

# BMI Calculation
BMI = weight / (height / 100) ** 2

print(f"Your BMI is: {round(BMI, 2)}")

# Categorize BMI
if BMI < 18.5:
    category = "You're Classified as UNDERWEIGHT"
elif 18.5 <= BMI <= 24.9:
    category = "You're Classified as NORMAL"
elif 25 <= BMI <= 29.9:
    category = "You're Classified as OVERWEIGHT"
else:
    category = "You're Classified as OBESE"

print(category)
```

```python
# Save results to file
save_results_to_file(name, BMI, category, weight, height)

# Plot the BMI
plot_bmi(BMI)
```
